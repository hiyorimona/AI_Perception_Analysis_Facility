```r
install.packages("broom", type="binary")
install.packages("tidyverse")
library(dplyr)
```

```r
# Reading both numerical and text df
df_main_text <-read.csv("survey_text_13_10.csv",header = TRUE,sep = ",")
df_main_num <- read.csv("survey_num_13_10.csv",header = TRUE,sep = ",")
```

```r
# dimension of df
dim(df_main_text)
dim(df_main_num)
```

```r
# filtering df for "Facility"
df_text <- df_main_text %>% filter(demo_domain %in% 'Facility')
df_num <- df_main_num %>% filter(demo_domain %in% 8) 

dim(df_text)
dim(df_num)
```

```r
# finding missing values
df_num[df_num == ""] <- NA
df_text[df_text == ""] <- NA

missing_sum_1 <- colSums(is.na(df_num))
missing_sum_2 <- colSums(is.na(df_text))
```

```r
# removing irrelevant columns df_num
columns_irrelevant_num <- names(missing_sum_1[missing_sum_1 == nrow(df_num)])
df_cleaned_num <- df_num[, !(names(df_num) %in% columns_irrelevant_num)]
df_num <- select(df_cleaned_num, -contains(c("Q","X")))
dim(df_num)

# removing irrelevant columns df_text
columns_irrelevant_text <- names(missing_sum_2[missing_sum_2 == nrow(df_text)])
df_cleaned_text <- df_text[, !(names(df_text) %in% columns_irrelevant_text)]
df_text <- select(df_cleaned_text, -contains(c("Q","X")))
dim(df_text)
```

```r
# subsetting for facilities df_num
df_num <- subset(df_num, select = c(12:ncol(df_num)-1))
dim(df_num)

# subsetting for facilities df_text
df_text <- subset(df_text, select = c(12:ncol(df_text)-1))
dim(df_text)
```

```r
head(df_text)
```

```r
# Assessing which columns should be kept 
sort(colSums(is.na(df_num)), decreasing = FALSE)
```

```r
names(df_num)
```

```r
# subsetting questions with less than 5 missing values from df_num
df_general_num <- subset(df_num, select = c(1:24))
df_subdomain_num <- subset(df_num, select = c(1:6,35:48))

# subsetting questions with less than 5 missing values from df_text
df_general_text <- subset(df_text, select = c(1:24))
df_subdomain_text <- subset(df_text, select = c(1:6,35:48))

names(df_general_text)
names(df_subdomain_text)
#dim(df_general_text)
```

```r
summary(df_general_text$demo_ai_know)
```

```r
head(df_subdomain_text,10)
```

```r
# removing all missing values from df_num
df_g_num_clean <- na.omit(df_general_num)
df_g_text_clean <- na.omit(df_general_text)

df_sub_num_clean <- na.omit(df_subdomain_num)
df_sub_text_clean <- na.omit(df_subdomain_text)


dim(df_g_num_clean)
dim(df_sub_text_clean)
```

```r
head(df_g_num_clean$acc_5,5)
head(df_sub_text_clean,5)
```

```r
# Changing values that are not correct
df_g_new <- df_g_num_clean %>%
  mutate(
    demo_ai_know = ifelse(demo_ai_know == 13, 5,
                          ifelse(demo_ai_know == 14, 4,
                          ifelse(demo_ai_know == 15, 3,
                          ifelse(demo_ai_know == 16, 2,
                          ifelse(demo_ai_know == 17, 1, 
                          demo_ai_know))))),
    acc_5 = ifelse(acc_5 == 15, 1,
                          ifelse(acc_5 == 16, 2,
                          ifelse(acc_5 == 17, 3,
                          ifelse(acc_5 == 18, 4,
                          ifelse(acc_5 == 19, 5,
                          acc_5))))))
```

```r
# exporting preprocessed data 
write.csv(df_g_new, file = "clean_survey_num_11_10.csv",row.names=FALSE)
```
